```python
Heatmap_Nochild=pd.read_csv('/content/gdrive/MyDrive/Application material/Applications/PMO/Population/heatmap2.csv',index_col="Unnamed: 0")
#Table Title: Resident Ever-Married Female Non-Students Aged 15 Years and Over by Age Group, Number of Children Born and Highest Qualification Attained (Census of Population 2010)
#drop rows with total in number of child
Heatmap_Nochild = Heatmap_Nochild.loc[Heatmap_Nochild["Number of Children"]!="Total"]
#get total number of people within each education level
Educationtotal=Heatmap_Nochild.groupby(["Education"]).sum()
#Heatmap_Nochild=Heatmap_Nochild.groupby(["Con_AgeGroup"]).sum()
Education_list=list(Heatmap_Nochild["Education"].unique())

##forming heatmap
 
#get table of no child
Heatmap_Nochild=Heatmap_Nochild.loc[Heatmap_Nochild["Parental status"]=="NONE"]
#remove row of number of child
Heatmap_Nochild=Heatmap_Nochild.drop(labels=["Number of Children","Parental status"],axis=1)
Heatmap_Nochild=pd.pivot_table(Heatmap_Nochild,values="value",index="Con_AgeGroup",columns="Education",aggfunc=np.sum)
print(Educationtotal['value'])
#rearrange rows
Heatmap_Nochild.index = pd.CategoricalIndex(Heatmap_Nochild.index, categories= ["> 50 years","40 - 49 years","30-39 years","20 - 29 years","< 20 years"])
Heatmap_Nochild.sort_index(level=0, inplace=True)
#rearrange columns
Heatmap_Nochild = Heatmap_Nochild[['Below Secondary', 'Secondary', 'Post-Secondary (Non-Tertiary)', 'Diploma* & Professional Qualification', 'University']]
Heatmap_Nochild=Heatmap_Nochild.drop(labels=['< 20 years'],axis=0)
#plot heatmap
for educ in Education_list:
  Heatmap_Nochild[educ]=Heatmap_Nochild[educ].div(Educationtotal['value'][educ]).mul(100)
ax = sns.heatmap(Heatmap_Nochild,cmap="BuPu")
ax.set_title('% of Singaporeans of specified highest edducation level with no child in 2010')
plt.ylabel('Age Group')
plt.xlabel('Education level')

plt.savefig('nochilds_heatmap.pdf', dpi=200,papertype='a10',format='pdf')
```
